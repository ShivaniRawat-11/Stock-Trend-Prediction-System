# Stock Market Trend Prediction - Exploratory Data Analysis (EDA)

This notebook demonstrates how to load, clean, engineer features (technical indicators), and inspect correlations for the Stock Market Trend Prediction System.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import custom pipeline modules
from src.data_preprocessing import load_stock_data, clean_data
from src.feature_engineering import create_features_and_target
from src.evaluation import set_premium_plot_style

set_premium_plot_style()
```

## 1. Load and Clean Data

```python
# Load AAPL stock data from 2020 to 2025
df_raw = load_stock_data(ticker='AAPL', start_date='2020-01-01', end_date='2025-01-01')
df_clean = clean_data(df_raw)
df_clean.head()
```

## 2. Feature Engineering & Technical Indicators

```python
# Generate SMA, EMA, RSI, MACD, returns, volatility, and shifted Target
df_feat = create_features_and_target(df_clean)
df_feat.head()
```

## 3. Visualize closing price and moving averages

```python
plt.figure(figsize=(12, 6))
plt.plot(df_feat.index[-250:], df_feat['Close'][-250:], label='Close Price', linewidth=2)
plt.plot(df_feat.index[-250:], df_feat['SMA_14'][-250:], label='SMA 14', linestyle='--')
plt.plot(df_feat.index[-250:], df_feat['SMA_50'][-250:], label='SMA 50', linestyle=':')
plt.title('AAPL Stock Price & Moving Averages (Last 250 Days)')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```

## 4. Feature Correlation Analysis

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df_feat.corr(), annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of All Engineered Features')
plt.show()
```
